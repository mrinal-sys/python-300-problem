### write a program to display count of digits in a given number 

```python
# Ask the user to enter a number
n = int(input("Enter any number: "))

# This variable will keep track of how many digits we have counted
count = 0

# Special case: if the number is 0, it still has 1 digit
if n == 0:
    count = 1
else:
    # Loop as long as the number is greater than 0
    while n > 0:
        count = count + 1         # Add 1 to the count
        n = n // 10               # Remove the last digit

# Print the total number of digits
print("Count of Digits:", count)
```
